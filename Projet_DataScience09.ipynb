### Fonction de nettoyage de pays 

```python
pays=af_Est.columns[0]
def NET(afrique, pays):
    list_Pays=list(afrique[pays])
    Pays_sans=[list_Pays[i].strip() for i in range(len(list_Pays))]
    Pays_Preopre=[Pays_sans[i].split()[0] for i in range(len(list_Pays))]
    return  Pays_Preopre

data=NET(af_Est, 'pays')
print(data)
```

## Calcul de la moyenne Ecart type et coefficient de Variation 

```python
Key= Africa.groupby('region')
Solution=Key['population'].agg([np.mean,np.std])
aff = Africa[['population','pays','region']].groupby(['region']).std()/Africa[['population','region']].groupby(['region']).mean()
Solution['CV']=aff['population']
Solution=Solution.reset_index(level = ['region'],inplace = False)
Solution=Solution.astype({"mean":int,"std":int,"CV":float})
Solution.columns = ['region','Mean','Standard deviation','Coeficient of variation'] 
Solution
```

# calcul de densité par Région

```python
all_region_d=Africa[['population','region','superficie']].groupby(['region']).sum()
```

```python
df=all_region_d.reset_index(level=['region'],inplace=False)
```

```python
h=(df['population']/df['superficie']).apply(lambda x: '%.1f' % x)
```

```python
df['Densité_Region'] = h
df
```
